**Понятие кредитного скоринга и скоринговых моделей**


Скоринг — это автоматизированная система оценки платежеспособности заемщика.

Она учитывает множество параметров: от кредитной истории до поисковых запросов клиента. Алгоритм оценивает эти данные и рассчитывает суммарный балл, на который банк ориентируется, принимая решение о выдаче кредита.

В 2018–2019 годах заметное увеличение потребительского кредитования стало одной из наиболее выделяющихся тенденций на российском рынке. Этот рост был обусловлен постепенным снижением процентных ставок в банковском секторе, что сделало выдачу разнообразных кредитных продуктов более привлекательной для банков. Повышенный спрос клиентов на кредиты в первую очередь был вызван периодом послекризисного накопления их потребительской активности.

В период с 2014 по 2016 годы российская экономика столкнулась с серьезными вызовами, и банковский сектор заметно сократил объемы выданных кредитов на душу населения. Эти события оказали значительное воздействие на рынок, но последующее восстановление и изменение процентных ставок способствовали восстановлению доверия кредитным продуктам

Кредитный риск – это риск возможных потерь в случае неспособности потенциальных заемщиков исполнять свои обязательства в соответствии со сроками и условиями кредитного договора. Кредитный риск — это ключевой вид риска, который регулярно возникает в процессах различных финансовых организаций

В процессе оценки кредитного риска применяются различные методы, как до, так и после выдачи кредита. Они включают:

1) Оценка при подаче заявки на кредит:

Application Scoring: Это оценка потенциального заемщика на основе предоставленных им данных при подаче заявки на кредит. Анализируются анкетные данные клиента для принятия решения о выдаче кредита.


2) Оценка после выдачи кредита:

Behavioral Scoring: Оценка основывается на анализе изменений в финансовой активности клиента, включая его кредитную историю, для определения вероятности дефолта.

Collection Scoring: Этот метод направлен на эффективное управление процессами взыскания просроченной задолженности, основываясь на анализе поведения клиента в отношении кредитного договора.

Fraud Scoring: Метод выявления потенциального мошенничества при подаче заявки на кредит, направленный на предотвращение финансовых потерь банка от недобросовестных клиентов.
Таким образом, методы кредитного скоринга различаются в зависимости от этапа кредитного процесса и целей оценки, будь то первоначальный анализ заемщика или последующее управление кредитным портфелем.

**Алгоритмы и критерии качества скоринговых моделей**


Скоринг помогает кредитной организации оценить риски в отношении заемщика и на начальном этапе отсеять тех, кто не соответствует банковским требованиям. Так банк выдает меньше рискованных кредитов, экономит время менеджеров и исключает ошибки, поскольку скоринговые системы автоматизированы. Заемщику скоринг позволяет подать заявку на кредит без визита в банк и получить предварительное одобрение за несколько минут.
Основными компонентами расчета кредитного скоринга являются:

1) Кредитная история: Информация о предыдущих кредитах, платежной дисциплине,
наличии просрочек и дефолтов.

2) Финансовое положение: Доход, имущество, обязательства и другие финансовые показатели.

3) Личные данные: Возраст, семейное положение, образование, тип занятости и другие социально-демографические характеристики.

4) Поведенческие факторы: Паттерны потребления, история покупок, поведение в социальных сетях и другие.

Для анализа вышеперечисленных факторов используются современный алгоритмы машиноого обучения:

Логистическая регрессия;

Деревья решений и случайные леса;

Градиентный бустинг;

Нейронные сети.



Кроме того, для построения надежных и точных моделей машинного обучения необходимо выбрать важные входные предикторы (информацию о должнике). Настоящая вычислительная работа сосредоточена на построении модели прогнозирования кредитного скоринга.

Процесс разработки модели включается в себя сбор данных, предобработку, разделение на обучающую и тестовую выборки, выбор модели, обучение модели, оценку производительности и внедрение модели.

Кредитным организациям крайне важно правильно оценивать, насколько хорошо модель предсказывает риск дефолта. Используются метрики, такие как ROC-AUC, точность, полнота, F1-скор.

Модели кредитного скоринга требуют регулярного обновления и мониторинга, чтобы адаптироваться к изменяющимся экономическим условиям и паттернам поведения клиентов.

Кредитный скоринг представляет собой сложную область, где аналитические методы сочетаются с экспертными знаниями для управления кредитными рисками в финансовом секторе.


В процессе создания моделей машинного обучения широко применяется методика разделения исходных данных на несколько подвыборок, чтобы обеспечить объективную оценку качества моделей. Обычно используются следующие подвыборки:

Обучающая выборка (Train, 60%-70%): Это основной набор данных, используемый для обучения модели.

Тестовая выборка (Test, 20%-30%): На этой части данных проводится тонкая настройка и первичная оценка качества модели для устранения переобучения.

Валидационная выборка (Out-of-time, примерно 10%): Этот набор данных используется для окончательного сравнения производительности модели с результатами, полученными на обучающей и тестовой выборках (не всегда применяется).


Одним из ключевых инструментов оценки эффективности моделей машинного обучения является матрица ошибок. Матрица ошибок отображает количество объектов, которые фактически принадлежат классу w_i, но были отнесены к классу w_j. Элементы, расположенные на главной диагонали матрицы, соответствуют верным классификациям, тогда как элементы вне диагонали указывают на ошибочные классификации.


Для бинарной классификации метка класса принимает, как правило, положительное, или отрицательное значение.
Для матрицы ошибок характерны 4 основных параметра:

True positive (TP);

False negative (FN);

False positive (FP);

True negative (TN).

Где P  - положительный класс объекта, N – отрицательный класс объекта . T  - класс предсказан правильно , а F - неправильно.



Поэтому, для скоринговых моделей обычно применимы следующие метрики качества:

Accuracy=(TP+TN)/(TP+FP+FN+TN) – доля правильно классифицированных объектов;

Precision=TP/(TP+FP) – доля правильно классифицированных плохих объектов среди всех наблюдений, отнесенных алгоритмом к плохим объектам;

Recall=TP/(TP+FN) – оценка способности алгоритма распознавать плохие объекты;

Negative Predictive Value=TN/(FN+TN) – доля правильно классифицированных хороших объектов среди всех наблюдений, отнесенных алгоритмом к хорошим объектам;

F1 score=(2*Precision*Recall)/(Precision+Recall) – среднее значение точности и полноты;

False Negative Rate=FN/(TP+FN) – доля плохих объектов, неправильно отнесенных к хорошим;

False Positive Rate=FP/(TN+FP) – доля хороших объектов, неправильно отнесенных к плохим.


ROC-кривая (Receiver Operating Characteristic) отображает изменение отношения доли Recall верно классифицированных плохих объектов в их общем количестве к доле False Positive Rate хороших объектов, неправильно отнесенных к плохим.
Площадь под ROC-кривой (AUC) есть не что иное, как показатель качества модели. Чем больше AUC, тем лучше разработанная модель.

**Логистическая регрессия:**

Основная концепция:
В отличие от линейной регрессии, предсказывающей непрерывные значения, логистическая регрессия используется для предсказания вероятности принадлежности объекта к определенному классу (например, да или нет). Математически это достигается с помощью логистической функции, известной как сигмоид, которая преобразует любое входное значение в значение между 0 и 1, тем самым представляя его как вероятность.

Математический аспект:
Центральный элемент модели — это логистическая функция: σ(z) = 1 / (1 + e^(-z)), где z - линейная комбинация входных признаков, взвешенная коэффициентами. Цель обучения модели — оптимизировать эти коэффициенты таким образом, чтобы модель наилучшим образом предсказывала класс объекта.

Обучение модели:
В процессе обучения логистической регрессии основной задачей является минимизация ошибки между предсказанными и реальными значениями. Для этого обычно используется метод максимального правдоподобия. Процесс обучения включает в себя использование оптимизационных алгоритмов, таких как градиентный спуск, для нахождения наилучших значений коэффициентов.

Оценка модели:
Для оценки качества модели логистической регрессии используются различные метрики, включая точность (accuracy), полноту (recall), точность (precision) и F1-скор. Также часто используется ROC-кривая и площадь под ней (AUC).


---



**Дерево решений:**

Ключевые принципы:

Простота и Интерпретируемость: Одно из главных преимуществ деревьев решений — простота понимания и интерпретации. Модели можно визуализировать, что делает их легко интерпретируемыми даже для пользователей без технического фона.

Автоматический отбор признаков: Деревья решений способны автоматически выбирать наиболее значимые признаки для классификации, что упрощает процесс моделирования и снижает риск переобучения.

Алгоритмы построения дерева:


Ключевые процессы в построении дерева:


1) Выбор признака: Определение атрибута, который лучше всего разделяет данные, часто основывается на мерах, таких как энтропия, прирост информации, индекс Джини.

2) Разбиение: Разделение данных на подмножества на основе этих атрибутов. Процесс продолжается рекурсивно.

3) Обрезка: Удаление частей дерева, которые не обеспечивают большой прирост информации, помогает избежать переобучения.


Преимущества и Недостатки:

Преимущества:

Простота понимания и интерпретации.
Не требует большой предобработки данных.
Способность обрабатывать как числовые, так и категориальные данные.

Недостатки:

Склонность к переобучению, особенно в случае большой глубины дерева.
Могут быть нестабильными, небольшие изменения в данных могут привести к сильно измененным деревьям.
Проблемы с моделированием некоторых типов зависимостей из-за их структуры решений.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, accuracy_score, f1_score, precision_score, recall_score

# Создание модели логистической регрессии
logreg = LogisticRegression()

# Удаляю фичи BAD, JOB, REASON из набора входных признаков
x_basic = df.drop(columns=["BAD", "JOB", "REASON"])
y = df["BAD"]

# Разделение данных на тренировочные и тестовые
x_basic_tr, x_basic_te, y_tr, y_te = train_test_split(x_basic, y, test_size=0.33, random_state=1)

# Обучение модели
logreg.fit(x_basic_tr, y_tr)

# Предсказание вероятностей и классов
y_pred_proba_logreg = logreg.predict_proba(x_basic_te)[:, 1]
y_pre = logreg.predict(x_basic_te)

# Вычисление ROC AUC
l_1 = roc_auc_score(y_te, y_pred_proba_logreg)

# Вычисление значений для ROC кривой
fpr, tpr, thresholds = roc_curve(y_te, y_pred_proba_logreg)

# Вычисление других метрик
a_1 = accuracy_score(y_te, y_pre)
f_1 = f1_score(y_te, y_pre, average="macro")
p_1 = precision_score(y_te, y_pre, average="macro")
r_1 = recall_score(y_te, y_pre, average="macro")

# Вывод результатов
print("Accuracy Score:", a_1)
print("F1 Score:", f_1)
print("Precision Score:", p_1)
print("Recall Score:", r_1)
print("ROC AUC:", l_1)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_te, y_pred_proba_logreg)
roc_auc = auc(fpr, tpr)

# Создание графика ROC кривой
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression')
plt.legend(loc="lower right")
plt.show()
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cnf_matrix = confusion_matrix(y_te, y_pre)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=["BAD"],
                      title='Confusion matrix - Logistic Regression Algorithm')

plt.show()
```
